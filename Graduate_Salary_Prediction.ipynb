### **Filling Null values with Mean(Null values in turn replace -1s in the data)**

```python
df=df.replace(-1,np.nan) #replace -1 with null values
cols_with_nan=[column for column in df.columns if df.isna().sum()[column]>0]
for column in cols_with_nan:
  df[column]=df[column].fillna(df[column].mean())  #mean imputation of null values
```

```python
#encoding
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
df.Gender=le.fit_transform(df.Gender)
df.Degree=le.fit_transform(df.Degree)
df.Specialization=le.fit_transform(df.Specialization)

df.head()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from xgboost import XGBRegressor
```

```python
x=df.drop('Salary',axis=1)
y=df['Salary']
```

```python
sc=StandardScaler()
x=sc.fit_transform(x)
```

```python
X_train,X_test,y_train,y_test=train_test_split(x,y,test_size=0.2,shuffle=True,random_state=0)
```

```python
xgb=XGBRegressor()
xgb.fit(X_train,y_train)
```

# Predict X_test data

```python
predictions=xgb.predict(X_test)
```

```python
predictions
```

```python
diff=y_test-predictions
```

# Visualize Difference

```python
plt.figure(figsize=(12,6))
sns.distplot(diff)
```

```python
xgb_r2_score=xgb.score(X_test,y_test)
print(xgb_r2_score)
```
